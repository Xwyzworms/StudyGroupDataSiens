```
import random
import math

choromosome_count = 50
generasi = 10000
pc = 0.65
pm = 0.4
```

```
def f(x1,x2):
    return (math.cos(x1)*math.sin(x2)- (x1/(x2**2+1)))
```

```
# real number encoding
def gen_population(count):
  populations = []
  for i in range(count):
    chromosome = {
        "x1": random.uniform(-1, 2),
        "x2": random.uniform(-1, 1)
    }
    populations.append(chromosome)
  
  return populations
```

```
#menggunakan fitness maksimasi
def fitness(populations):
  result = []
  for i in range(choromosome_count):
    c = populations[i]
    h = f(c['x1'], c['x2'])
    a = 0.001
    fit = 1/(h+a)
    result.append({
        'x1': c['x1'],
        'x2': c['x2'],
        'index': i,
        'h': h,
        'fit': fit
    })
  return result
```

```
#rank selection
def get_parrent(populations, fit_res):
  sorted_fit = sorted(fit_res, key=lambda x: x['fit']) 
  parent1 = populations[sorted_fit[0]['index']]
  parent2 = populations[sorted_fit[1]['index']]
  return parent1, parent2
```

```
#one point crossover
def crossover(parrent):
  p1, p2 = parrent
  child1 = []
  child2 = []
  child1={
    "x1": p1['x1'],
    "x2": p2['x2']
    }
  child2={
    "x1": p2['x1'], 
    "x2": p1['x2']
    }
  return(child1,child2)

#mutation tingkay 1 bit
def mutation(child):
  p1, p2 = child
  child1 = []
  child2 = []
  i = random.randint(0, 1)
  if i==0:
      child1={
        'x1': random.uniform(-1, 2),
        'x2': p2['x2']
        }
      child2={
        'x1': random.uniform(-1, 2), 
        'x2': p1['x2']
        }
  else:
      child1={
        'x1': p1['x1'],
        'x2': random.uniform(-1, 1)
        }
      child2={
        'x1': p2['x1'], 
        'x2': random.uniform(-1, 1)
        }
  return(child1,child2)
```

```
def survivor_selecton(populasi, fit_res, child):
  sorted_fit = sorted(fit_res, key=lambda x: x['fit'])
  index_mati = sorted_fit[len(sorted_fit)-2]['index']
  populasi.append(child)
  return populasi
```

```
populations = gen_population(choromosome_count)
fit_res = fitness(populations)
parent = get_parrent(populations, fit_res)
child = crossover(parent)
child = mutation(child)
populations = survivor_selecton(populations, fit_res, child)
for i in range(generasi):
    fit_res = fitness(populations)
    parent = get_parrent(populations, fit_res)
    b = random.uniform(0.0, 1.0)
    if b <= pc:
        child = crossover(parent)
    c = random.uniform(0.0, 1.0)
    if c <= pm:
        child = mutation(child)
    populations = survivor_selecton(populations, fit_res, child)
print(fitness(populations))
```
